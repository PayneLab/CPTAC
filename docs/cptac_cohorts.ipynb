```python
""" Transcriptomics
This strategy ratios each sample's abundance to a pooled reference in TMT. The ratio is 
then log transformed. Therefore positive values indicate a measurement 
higher than the pooled reference; negative values are lower than the pooled reference.
"""
dataf = g.get_transcriptomics()
print(dataf[0:21])
```

```python
#How to choose certain rows and columns
samples = ["S001","S003","S016"]
proteins = ["A1BG","PTEN","TP53"]
dataf.loc[samples, proteins]
```

```python
#Search for expression greater than 0 for a specific gene
brca_positive = dataf.loc[dataf["BRCA1"] > 0]
brca_positive.head()
```
